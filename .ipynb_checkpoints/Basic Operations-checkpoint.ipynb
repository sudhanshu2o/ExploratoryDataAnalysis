```python
#NORMALIZATION of Values of Age
from sklearn.preprocessing import MinMaxScaler
```

```python
data_age_norm=scale.transform(data_age)
```

```python
data["Age"]=data_age_norm
data.head()
```

```python
#NORMALIZATION of Values of FARE
data_Fare=data["Fare"].values.reshape(-1,1)
scale=MinMaxScaler()
scale=scale.fit(data_Fare)
data_Fare_norm=scale.transform(data_Fare)
data["Fare"]=data_Fare_norm
data.head()
```

# PAIRPLOT

```python
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data.isnull().any()
```

```python
data["Age"]=data["Age"].fillna(data["Age"].mean())
```

```python
sns.pairplot(data)
```

```python
data['Sex']=data['Sex'].map({'male':0 , 'female':1})
```

```python
#correlation mapping
cols=["Survived","Age","Pclass","Fare","Sex","PassengerId"]
data_corr=data[cols].corr()
print(data_corr)
```

```python
sns.heatmap(data_corr)
```

```python
#3D graph
ax=plt.axes(projection="3d")
ax.scatter(data["Age"],data["Sex"],data["Fare"])
plt.show()
```

```python
import pandas_bokeh
data.plot_bokeh(kind="Scatter",x=data["Survived"]),
data.plot_bokeh(kind="Scatter",x=data["Age"]),
data.plot_bokeh(kind="Scatter",x=data["Fare"])
```
